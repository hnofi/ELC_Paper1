```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import healpy as hp
import astropy.io.fits as fits
from matplotlib import cm
import matplotlib.colors as mcolors
```

```python
# Input cleaned maps from initial cleaning fit
initial_70_map="INITIAL_Maps/Initial_Cleaned_70GHz.fits"
initial_70=hp.fitsfunc.read_map(initial_70_map,field=[0])

initial_94_map="INITIAL_Maps/Initial_Cleaned_94GHz.fits"
initial_94=hp.fitsfunc.read_map(initial_94_map,field=[0])

initial_100_map="INITIAL_Maps/Initial_Cleaned_100GHz.fits"
initial_100=hp.fitsfunc.read_map(initial_100_map,field=[0])

initial_143_map="INITIAL_Maps/Initial_Cleaned_143GHz.fits"
initial_143=hp.fitsfunc.read_map(initial_143_map,field=[0])
```

```python
figsize = (25, 10)
fig = plt.figure(figsize=figsize)
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams.update({'font.size': 40}) 

hp.cartview(initial_70, cmap='rainbow', latra=[-10, 10], min=-0.2, max=0.2, sub=411, title="",cbar=False,badcolor='black')
plt.title("70 GHz", loc='left')

hp.cartview(initial_94, cmap='rainbow', latra=[-10, 10], min=-0.2, max=0.2, sub=412, title="",cbar=False,badcolor='black')
plt.title("94 GHz", loc='left')

hp.cartview(initial_100, cmap='rainbow', latra=[-10, 10], min=-0.2, max=0.2, sub=413, title="",cbar=False,badcolor='black')
plt.title("100 GHz", loc='left')

hp.cartview(initial_143, cmap='rainbow', latra=[-10, 10], min=-0.2, max=0.2, sub=414, title="",cbar=False,badcolor='black')
plt.title("143 GHz", loc='left')

cbar_ax = fig.add_axes([0.15, -0.03, 0.7, 0.04]) 
norm = mcolors.Normalize(vmin=-0.2, vmax=0.2)
sm = cm.ScalarMappable(cmap='rainbow', norm=norm)
cbar = fig.colorbar(sm, cax=cbar_ax, orientation='horizontal')
cbar.set_label(r'mK$_{\rm CMB}$')

plt.show()
```
